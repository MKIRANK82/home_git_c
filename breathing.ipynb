```python
import tkinter as tk
from tkinter import messagebox
import time
import datetime
import winsound  # This is for Windows; you can use another library for non-Windows systems


def play_sound(frequency, duration):
    """Plays a beep sound. Frequency and duration in milliseconds."""
    print(f"in play sound {datetime.datetime.now()}")
    winsound.Beep(frequency, duration)


def start_breathing():
    try:
        inhale_duration = int(inhale_var.get())
        hold_duration = int(hold_var.get())
        exhale_duration = int(exhale_var.get())
        hold_exhale_duration = int(hold_exhale_var.get())
        total_duration = int(total_time_var.get())

        start_time = time.time()
        elapsed_time = 0

        while elapsed_time < total_duration:
            # Inhale phase
            play_sound(500, 500)  # Sound for inhale (frequency: 500Hz, duration: 500ms)
            time.sleep(inhale_duration)

            # Hold phase after inhale
            play_sound(600, 500)  # Sound for hold (frequency: 600Hz, duration: 500ms)
            time.sleep(hold_duration)

            # Exhale phase
            play_sound(400, 500)  # Sound for exhale (frequency: 400Hz, duration: 500ms)
            time.sleep(exhale_duration)

            # Hold phase after exhale
            play_sound(300, 500)  # Sound for hold after exhale (frequency: 300Hz, duration: 500ms)
            time.sleep(hold_exhale_duration)

            # Update elapsed time
            elapsed_time = time.time() - start_time

        messagebox.showinfo("Breathing Exercise", "Breathing session completed!")
    
    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid integers for the durations.")


# Create the GUI window
root = tk.Tk()
root.title("Breathing Exercise")

# Create input fields for each phase
tk.Label(root, text="Inhale Duration (seconds):").grid(row=0, column=0)
inhale_var = tk.StringVar()
tk.Entry(root, textvariable=inhale_var).grid(row=0, column=1)

tk.Label(root, text="Hold Duration (seconds):").grid(row=1, column=0)
hold_var = tk.StringVar()
tk.Entry(root, textvariable=hold_var).grid(row=1, column=1)

tk.Label(root, text="Exhale Duration (seconds):").grid(row=2, column=0)
exhale_var = tk.StringVar()
tk.Entry(root, textvariable=exhale_var).grid(row=2, column=1)

tk.Label(root, text="Hold After Exhale Duration (seconds):").grid(row=3, column=0)
hold_exhale_var = tk.StringVar()
tk.Entry(root, textvariable=hold_exhale_var).grid(row=3, column=1)

tk.Label(root, text="Total Time (seconds):").grid(row=4, column=0)
total_time_var = tk.StringVar()
tk.Entry(root, textvariable=total_time_var).grid(row=4, column=1)

# Start button to begin the breathing exercise
start_button = tk.Button(root, text="Start Breathing", command=start_breathing)
start_button.grid(row=5, columnspan=2)

# Run the GUI main loop
root.mainloop()
```

```python
import re
def get_volume(record):
    match = re.search(r'vol(?:ume)?\s*(\d+)', record, re.IGNORECASE)
    # print(match)
    return int(match.group(1)) if match else float('inf')  # 'inf' if no volume is found, to push it to the end

# Sort data by volume number
with open("breathing level 4.txt","r") as f:
    datas=f.readlines();
data=[]
for item in datas:
    if "level 4" in item.lower():
        data.append(item)
sorted_datas = sorted(data, key=get_volume)
sorted_data = list(dict.fromkeys(sorted_datas))

# Display sorted results
file_data=""
for record in sorted_data:
    # print(record)
    file_data=file_data+" "+record
print(file_data)

with open("breathing level 4 sorted.txt","w") as f:
    f.write(file_data)
    
```
